```sql
-- Gold BLS employment-projection fact view
-- Grain: one row per SOC code and BLS occupation type.

CREATE OR REPLACE VIEW workforce_analytics.gold.fact_occupation_projection
COMMENT 'BLS 2024-2034 occupation projections with readable employment counts, growth direction, wages, education, and training.'
AS

SELECT
    soc_code,
    occupation_title,
    occupation_type,

    2024 AS projection_base_year,
    2034 AS projection_target_year,

    employment_base_year_thousands,
    employment_projected_year_thousands,
    employment_change_thousands,
    annual_openings_thousands,

    CAST(
        ROUND(
            employment_base_year_thousands * 1000,
            0
        ) AS BIGINT
    ) AS employment_base_year,

    CAST(
        ROUND(
            employment_projected_year_thousands * 1000,
            0
        ) AS BIGINT
    ) AS employment_projected_year,

    CAST(
        ROUND(
            employment_change_thousands * 1000,
            0
        ) AS BIGINT
    ) AS employment_change,

    CAST(
        ROUND(
            annual_openings_thousands * 1000,
            0
        ) AS BIGINT
    ) AS annual_openings,

    employment_change_percent,

    CASE
        WHEN employment_change_percent > 0
            THEN 'growing'
        WHEN employment_change_percent < 0
            THEN 'declining'
        WHEN employment_change_percent = 0
            THEN 'no_change'
        ELSE 'unknown'
    END AS growth_direction,

    employment_distribution_base_percent,
    employment_distribution_projected_percent,
    percent_self_employed,

    median_annual_wage,
    median_annual_wage_raw,
    median_annual_wage_qualifier,

    typical_entry_education,
    related_work_experience,
    on_the_job_training,

    source_file_path,
    run_id,
    processed_at

FROM workforce_analytics.silver.bls_employment_projections

WHERE soc_code RLIKE '^[0-9]{2}-[0-9]{4}$';
```
