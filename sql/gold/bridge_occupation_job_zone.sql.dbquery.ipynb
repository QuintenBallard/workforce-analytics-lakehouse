```sql
-- Gold occupation-to-job-zone bridge
-- Grain: one row per detailed O*NET occupation.

CREATE OR REPLACE VIEW workforce_analytics.gold.bridge_occupation_job_zone
COMMENT 'O*NET Job Zones with readable preparation-level labels and detailed occupation titles.'
AS

SELECT
    job_zone.onet_soc_code,
    job_zone.soc_code,
    occupation.occupation_title,
    occupation.description AS occupation_description,

    job_zone.job_zone,

    CASE job_zone.job_zone
        WHEN 1 THEN 'little_or_no_preparation'
        WHEN 2 THEN 'some_preparation'
        WHEN 3 THEN 'medium_preparation'
        WHEN 4 THEN 'high_preparation'
        WHEN 5 THEN 'extensive_preparation'
        ELSE 'unknown'
    END AS job_zone_label,

    job_zone.source_file_path,
    job_zone.run_id,
    job_zone.processed_at

FROM workforce_analytics.silver.onet_job_zones AS job_zone

LEFT JOIN workforce_analytics.silver.onet_occupations AS occupation
    ON job_zone.onet_soc_code = occupation.onet_soc_code;
```
