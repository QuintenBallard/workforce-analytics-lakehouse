```sql
-- Gold occupation-to-software bridge
-- Grain: one row per O*NET occupation, software item, and O*NET element.

CREATE OR REPLACE VIEW workforce_analytics.gold.bridge_occupation_software
COMMENT 'O*NET occupation-to-software relationships including hot-technology and in-demand indicators.'
AS

SELECT DISTINCT
    software.onet_soc_code,
    software.soc_code,
    occupation.occupation_title,

    software.software_name,
    software.element_id,
    software.element_name,

    software.hot_technology,
    software.in_demand,

    software.source_file_path,
    software.run_id,
    software.processed_at

FROM workforce_analytics.silver.onet_software_skills AS software

LEFT JOIN workforce_analytics.silver.onet_occupations AS occupation
    ON software.onet_soc_code = occupation.onet_soc_code

WHERE software.software_name IS NOT NULL;
```
